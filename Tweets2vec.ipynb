```python
import json
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from flatten_json import flatten
import os
import re
import numpy as np
```

```python
final_global=pd.DataFrame()
```

```python
def tweet_cleaning(data):
    
    #flatenning of json file
    dic_flattened = (flatten(d) for d in data)
    df = pd.DataFrame(dic_flattened)
    
    # filtering english tweets
    X=df.loc[df['lang'] == 'en']
    Y=X.select(lambda col: col.startswith('entities_hashtags_'), axis=1).filter(like='text')
    
    #dropping na columns
    Y.dropna(axis=1,how='all',inplace=True)
    result = pd.concat([Y, X['text'],X['created_at']], axis=1)
    
    # drop rows where there is not even one hashtags
    result.dropna(axis=0,subset=['entities_hashtags_0_text'],inplace=True)
    
    # cleaning tweets - removing hyperlinks,#,\n
    result['text'] =result['text'].str.replace('http\S+','',case=False)
    result['text'] =result['text'].str.replace('#\S+','',case=False)
    result['text'] =result['text'].str.replace("\\n",'',case=False)
    #result['text'] =result['text'].str.replace("\d",'',case=False)
    
    # deleting Retweeted tweets
    df=result[result.text.str.contains("RT @") == False]
    
    # resetting the index
    df.reset_index(drop=True,inplace=True)
    df_2=df[['entities_hashtags_0_text','text']]
    
    
    #making dataframe from entities_hashtags_1_text
    df_3=df[df['entities_hashtags_1_text'].notnull()==True][['entities_hashtags_1_text','text']]
    df_3.reset_index(drop=True,inplace=True)
    df_3.columns=['entities_hashtags_0_text','text']
    
    #making dataframe from entities_hashtags_2_text
    if 'entities_hashtags_2_text' in df.columns:
        df_4=df[df['entities_hashtags_2_text'].notnull()==True][['entities_hashtags_2_text','text']]
        df_4.reset_index(drop=True,inplace=True)
        df_4.columns=['entities_hashtags_0_text','text']
    else:
        df_4=pd.DataFrame()
    
    #making dataframe from entities_hashtags_3_text
    if 'entities_hashtags_3_text' in df.columns:
        df_5=df[df['entities_hashtags_3_text'].notnull()==True][['entities_hashtags_3_text','text']]
        df_5.reset_index(drop=True,inplace=True)
        df_5.columns=['entities_hashtags_0_text','text']
        
    #making dataframe from entities_hashtags_4_text
    if 'entities_hashtags_4_text' in df.columns:
        df_6=df[df['entities_hashtags_4_text'].notnull()==True][['entities_hashtags_4_text','text']]
        df_6.reset_index(drop=True,inplace=True)
        df_6.columns=['entities_hashtags_0_text','text']
    else:
        df_6=pd.DataFrame()
        
    #making dataframe from entities_hashtags_5_text
    if 'entities_hashtags_5_text' in df.columns:
        df_7=df[df['entities_hashtags_5_text'].notnull()==True][['entities_hashtags_5_text','text']]
        df_7.reset_index(drop=True,inplace=True)
        df_7.columns=['entities_hashtags_0_text','text']
    else:
        df_7=pd.DataFrame()
    
    # Combining all the dataframes
    df_2=df_2.append([df_3,df_4,df_5,df_6,df_7])
    df_2.reset_index(drop=True,inplace=True)
  
    
    #print(df_2)
    return df_2
    
    
```

```python
# directory path of folders containing files or folder containing subfolders which contains json files
source_1 =pwd()

# looping through directory sub files/folders
for root, dirs, filenames in os.walk(source_1):
    for f in filenames:
        fullpath = os.path.join(root, f)
        print(fullpath)
        tweets_data=[]
        tweets_file=open(fullpath,'r')
        
        for line in tweets_file:
            try:
                tweet=json.loads(line)
                tweets_data.append(tweet)
            except:
                continue
                    
        #from IPython.core.debugger import Tracer; Tracer()() 
        final_local=tweet_cleaning(tweets_data)
         
        
        # final_global is our output of this code, it is a dataframe containing two columns: hashtags and tweets.
        final_global=final_global.append(final_local,ignore_index=True)
        #from IPython.core.debugger import Tracer; Tracer()()
```
